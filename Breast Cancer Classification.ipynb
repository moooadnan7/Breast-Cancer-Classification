```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from keras import Sequential
from keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv('/content/data.csv')
data
```

```python
print("Number of malignant",data['diagnosis'].value_counts()[0])
print("Number of benign",data['diagnosis'].value_counts()[1])
```

```python
data.describe()
```

```python
data.columns
```

```python
data.drop(['id','Unnamed: 32'],axis=1,inplace=True)
data
```

```python
lb=LabelBinarizer()
data['diagnosis']=lb.fit_transform(data['diagnosis'])
data
```

```python
x=data.drop(['diagnosis'],axis=1)
y=data['diagnosis']
```

```python
st=MinMaxScaler()
x=st.fit_transform(x)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = Sequential()
model.add(Dense(30, activation='relu', input_dim=30))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=50,validation_split=0.2 )
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

R2 = r2_score(y_test, y_pred)
print("R2 Score=",R2 )
```
